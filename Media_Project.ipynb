```python
ds = ds.drop("Unnamed: 0.1", axis=1) 
ds.rename(columns={'Unnamed: 0':'S.no'},inplace=True)
ds.columns
print(ds.columns)
```

```python
import matplotlib.pyplot as py
import seaborn as sea
```

```python
num_col=ds.select_dtypes(include="number").columns
```

```python
col=['S.no','Retweets', 'Likes', 'Year', 'Month', 'Day', 'Hour']
```

```python
for columns in col:
    sea.histplot(ds[col],bins=10,kde=True)
    py.title(col)
    py.show()
```

```python
import scipy.stats as zscore
zscore=ds[['S.no','Retweets', 'Likes', 'Year', 'Month', 'Day', 'Hour']]
outliers=(zscore > -3)|(zscore < 3)
total_outliers=outliers.sum()
print(total_outliers)
```

```python
col=['S.no','Retweets', 'Likes', 'Year', 'Month', 'Day', 'Hour']
original_stats={}
```

```python
for col in num_col:
    mean_val=ds[col].mean()
    ds[col]=ds[col].fillna(mean_val)
    std_val=ds[col].std()
    original_stats[col]=(mean_val,std_val)
    ds[col]=ds[col].clip(lower = mean_val -3 * std_val,upper = mean_val + 3 * std_val)
    print(f"clipped value in {col}")
```

```python
for col in num_col:
    mean_val,std_val=original_stats[col]
    lower_bound=mean_val-3*std_val
    upper_bound=mean_val+3*std_val
    outliers_remaining=ds[(ds[col] > upper_bound)|(ds[col] < lower_bound)].shape[0]
    if outliers_remaining==0:
        print(f"remaining value {outliers_remaining} in {col}")
    else:
        print("Outliers is there")
    
```
